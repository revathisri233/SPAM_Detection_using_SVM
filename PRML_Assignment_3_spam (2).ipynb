```python
import os
os.chdir(r"C:\Users\user\Downloads")  # Path where files are present

def funct(loc):
    
    import csv
    file = open(loc)
    csvreader = csv.reader(file)
    header = next(csvreader)
    file.close()
    
    # Importing the data ( Dataset taken from kaggle)
    data = pd.read_csv(loc)
    data.rename(columns = {'v1':'Label', 'v2':'sms'}, inplace = True)
    data.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis = 1, inplace = True)
    data

    # Understanding the data
    data['Label'].value_counts()

    # Cleaning the data
    # Remove punctuations from the emails
    data['sms'] = data['sms'].str.replace('[^\w\s]','')

    # converting strings to lower case
    data['sms'] = data['sms'].str.lower()
    data.head()

    # creating a dictionary of words from sms
    data['sms'] = data['sms'].str.split()
    words = []
    for sms in data['sms']:
       for word in sms:
          words.append(word)

    words = list(set(words))
    len(words)

    # we have a total 9483 unique words in the data
    # Transforming the data into the dataframe we need
    # feature extraction-feature is the number of times a word repeats in a mail
    words_per_sms = {unique_word: [0] * len(data['sms']) for unique_word in words}

    for index, sms in enumerate(data['sms']):
       for word in sms:
          words_per_sms[word][index] += 1

    word_counts = pd.DataFrame(words_per_sms)
    word_counts.head()

    # adding Label and sms columns from training set
    global data_clean
    data_clean = pd.concat([data, word_counts], axis=1)
    data_clean.head()

    data_X = data_clean.iloc[:, 2:]
    data_y = data_clean.Label
    print(data_X.shape)
    print(data_y.shape)
    print(data_X)
 
loc=r"C:\Users\user\Downloads\spam.csv"
funct(loc)

# Split the dataset into train and test data
# Randomize the dataset
data_randomized = data_clean.sample(frac=1, random_state=1)

# Calculate index for split
training_test_index = round(len(data_randomized) * 0.8)

# Split into training and test sets
training_set = data_randomized[:training_test_index].reset_index(drop=True)
test_set = data_randomized[training_test_index:].reset_index(drop=True)

print(training_set.shape)
print(test_set.shape)

# Analysing the percentage of spam and ham messages in test and training data and both seem to be nearly eaual
training_set['Label'].value_counts(normalize=True)
test_set['Label'].value_counts(normalize=True)

train_X = training_set.iloc[:, 2:]
train_y = training_set.Label
print(train_X.shape)
print(train_y.shape)
print(train_X)
print(train_y)

test_X = test_set.iloc[:, 2:]
test_y = test_set.Label
print(test_X.shape)
print(test_y.shape)
print(test_X)
print(test_y)


# import support vector classifier
# "Support Vector Classifier"
from sklearn.svm import SVC 
clf = SVC(kernel='linear')
clf1 = SVC(kernel = 'rbf', random_state = 0)
 
# fitting x samples and y classes
clf.fit(train_X,train_y)
clf1.fit(train_X,train_y)

print(clf.score(test_X,test_y))
print(clf1.score(test_X,test_y))

import os
os.chdir(r"C:\Users\user\Downloads\PRML_assignment3\test")  # Path where files are present

def funct2(loc):
    
    import csv
    file = open(loc)
    csvreader = csv.reader(file)
    header = next(csvreader)
    file.close()
    
    # Importing the data ( Dataset taken from kaggle)
    dt = pd.read_csv(loc)
    # Understanding the data
    dt['Label'].value_counts()

    # Cleaning the data
    # Remove punctuations from the emails
    dt['sms'] = dt['sms'].str.replace('[^\w\s]','')

    # converting strings to lower case
    dt['sms'] = dt['sms'].str.lower()
    
    # creating a dictionary of words from sms
    dt['sms'] = dt['sms'].str.split()
    words = []
    for sms in dt['sms']:
       for word in sms:
          words.append(word)

    words = list(set(words))
    len(words)
    

    # we have a total 9483 unique words in the data
    # Transforming the data into the dataframe we need
    # feature extraction-feature is the number of times a word repeats in a mail
    words_per_sms = {unique_word: [0] * len(dt['sms']) for unique_word in words}

    for index, sms in enumerate(dt['sms']):
       for word in sms:
          words_per_sms[word][index] += 1

    word_counts = pd.DataFrame(words_per_sms)
    word_counts.head()

    # adding Label and sms columns from training set
    dt_clean = pd.concat([dt, word_counts], axis=1)
    dt_clean.head()
    
    global dt_X
    dt_X = dt_clean.iloc[:, 2:]
    dt_y = dt_clean.Label
    print(dt_X.shape)
    print(dt_y.shape)
    print(dt_X)
 
loc=r"C:\Users\user\Downloads\PRML_assignment3\test\email1.csv"
funct2(loc)

y_predict = clf.predict(dt_X)

print(y_predict)
 
```
